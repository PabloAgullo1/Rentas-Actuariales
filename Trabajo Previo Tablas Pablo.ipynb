```python
TablaFiltrada_df = TablaMortalidad_df.drop(labels = ["x+t", "Unnamed: 3", "Probabilidad anual de muerte", 'Unnamed: 5', 'Unnamed: 6' ], axis = 1)
```

```python
t_list = []
for i in range(0, len(TablaFiltrada_df["qx+t, tabla base"])):
    t_list.append(i)
TablaFiltrada_df.insert(loc = 0, column = "t", value = t_list)
```

```python
import math
def q_x(qx, mejora, t):
    '''
    La función q_x(x, mejora) recibe la probabilidad de muerte qx de la tabla de mortalidad y el factor de mejora de ésta
    Devuelve la probabilidad de muerte qx ajustada por el factor de mejora.
    '''
    return qx*math.exp(-mejora*t)
```

```python
g = int(input("Introduce el año de nacimiento x: "))

x_mas_t = 2000 - g

print("La edad que tenia esta persona en el año 2000 es: ", x_mas_t, " años")
```

```python
TablaResultados_df = TablaFiltrada_df.copy()
```

```python
TablaResultados_df.head()
```

```python
x_mas_t_list = []
for i in range(0, len(TablaResultados_df["qx+t, tabla base"])):
    x_mas_t_list.append(x_mas_t + i)
TablaResultados_df.insert(loc = 1, column="x+t", value = x_mas_t_list)
```

```python
Tabla_Mortalidad_final = TablaResultados_df.merge(TablaFiltrada_df, left_on= "x+t", right_on= "t", how = "left")
```

```python
Tabla_Mortalidad_final = Tabla_Mortalidad_final.drop(labels = ["t_y", "qx+t, tabla base_x", "λx+t_x" ], axis = 1)
Tabla_Mortalidad_final.head()
```

```python
Calculos = []
for i in range(0, len(Tabla_Mortalidad_final["qx+t, tabla base_y"])):
    Calculos.append(q_x(Tabla_Mortalidad_final["qx+t, tabla base_y"][i],
                        Tabla_Mortalidad_final["λx+t_y"][i], 
                        Tabla_Mortalidad_final["t_x"][i] ))
    
    #Tengo que buscar en la tabla de mortalidad filtrada la qx a partir del valor x+t como primer parámetro
    #de igual manera buscar el factor de mejora como segundo parámetro
    #como tercer parámetro es suficiente con usar el valor t que siempre es el mismo independientemente de la generación (g)
    
Tabla_Mortalidad_final.insert(loc=4, column = "q_x", value = Calculos)
```

```python
Tabla_Mortalidad_final.rename(columns={"t_x": "t", "qx+t, tabla base_y": "qx+t, tabla base", "λx+t_y": "λx+t"}, inplace=True)
```

```python
Mortality_Table = Tabla_Mortalidad_final.drop(labels = ["qx+t, tabla base", "λx+t"], axis = 1)
Mortality_Table.head()
```

```python
Mortality_Table['lx'] = 0  
Mortality_Table.loc[0, 'lx'] = 1_000_000  

# Calculamos lx y dx fila por fila
for i in range(1, len(Mortality_Table)):
    Mortality_Table.loc[i, 'lx'] = Mortality_Table.loc[i-1, 'lx'] * (1 - Mortality_Table.loc[i-1, 'q_x'])

Mortality_Table['dx'] = Mortality_Table['lx'] * Mortality_Table['q_x']
Mortality_Table
```

```python
def tpx(x, t = 1):
    '''
    La función tpx(x+t, t) recibe la edad x y el año t.
    Devuelve la probabilidad de que un individuo con x años recien cumplidos
    sobreviva x + t años más.
    
    Si no se especifica el año t, se asume que es 1.
    '''
    l_x = None
    l_x_mas_t = None
    
    for i in range(0, len(Mortality_Table["x+t"])):
        if Mortality_Table["x+t"][i] == x:
            l_x = Mortality_Table["lx"][i]
    
    for i in range(0, len(Mortality_Table["x+t"])):
        if Mortality_Table["x+t"][i] == x + t:
            l_x_mas_t = Mortality_Table["lx"][i]
    
    if l_x is None or l_x_mas_t is None:
        raise ValueError("Edad x o x+t no encontrada en la tabla de mortalidad")
    
    return l_x_mas_t / l_x

def tqx(tpx):
    return 1 - tpx

```

```python
edad = int(input("Introduce la edad x: "))
tiempo = int(input("Introduce el número de años a sobrevivir t: "))

supervivencia = tpx(edad, tiempo)

print("La probabilidad de sobrevivir", tiempo, " años es:", supervivencia) #43 años 22 sobrevivir, 0,91481559
```
